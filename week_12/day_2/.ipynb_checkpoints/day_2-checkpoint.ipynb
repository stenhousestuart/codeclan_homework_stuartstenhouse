## 3.

```python
# Select Required Columns and Save To New Data Frame
drinks = drinks_content[["Beverage_category", "Beverage", "Beverage_prep", "Calories"]]
drinks
```

```python
# Check for Any Missing Values 
drinks.isnull().sum()
```

## 3.

```python
drinks["calorie_diff"] = drinks["Calories"] - 135
drinks
```

## 5.

```python
drinks.groupby("Beverage_category").mean("Calories")
```

## 6.

```python
# Calculate mean calories calculated across all drinks.
mean_calories_all = drinks["Calories"].mean()
mean_calories_all
```

```python
# Create a new dataframe grouped by `Beverage_prep` and containing the `mean_calories` for each group.
beverage_prep_grouped = drinks.groupby("Beverage_prep")["Calories"].mean().reset_index(name = "mean_calories").sort_values("mean_calories")
beverage_prep_grouped
```

```python
# Create a new variable called `more_than_avg_calories`
beverage_prep_grouped["more_than_avg_calories"] = beverage_prep_grouped["mean_calories"] > mean_calories_all
```

```python
# Filter `more_than_avg_calories` to display only the `Beverage_prep` groups with higher than average calories.
beverage_prep_grouped[beverage_prep_grouped["more_than_avg_calories"] == True]
```
