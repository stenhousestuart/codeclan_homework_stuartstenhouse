```python
import pandas as pd
```

```python
from janitor import clean_names
```

## 1.

```python
cards = pd.read_csv("data/CardBase.csv").clean_names()
customers = pd.read_csv("data/CustomerBase.csv").clean_names()
fraud = pd.read_csv("data/FraudBase.csv").clean_names()
transactions = pd.read_csv("data/TransactionBase.csv").clean_names()
```

## 2.

```python
cards_customers = cards.merge(customers, how = "left", on = "cust_id")
cards_customers
```

## 3.

The row number after the join is the same as the row number of `transactions`. This means that there are no entries in the `fraud` data-frame that do not have a matching `transaction_id` in the `transactions` data-frame.

```python
transaction_fraud = transactions.merge(fraud, how = "outer", on = "transaction_id")
transaction_fraud
```

## 4.

```python
# Step 1. Merge Both Tables Using An Inner Join (Only Rows With Matches Are Kept, So If A Card Does Not )
cards_transactions = cards.merge(transactions, how = "inner", left_on = "card_number", right_on = "credit_card_id")
cards_transactions
```

```python
# Step 2. Search To See if the `card_number` from `cards` is in `card_number` in the newly merged `cards_transactions`
cards["card_number"].isin(cards_transactions["card_number"])
```

```python
# Step 3. Subset `cards_transactions` using The Filter from Step 2.
```

From the result, the number of rows matches the number of rows in `cards`. This suggests that all cards have made at least one transaction and some have made more than one, since the merged `cards_transactions` table had 1000 rows.

```python
cards[cards["card_number"].isin(cards_transactions["card_number"])]
```
